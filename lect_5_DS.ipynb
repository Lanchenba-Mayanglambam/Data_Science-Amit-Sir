```python
import numpy as np
```

```python
a = np.random.randint(1,90,size=(5,5))
# print(a)
# b = a.reshape(5,5)
# print(b)
# for i in range(5):
#   mean = np.mean(b[i])
#   print("mean of row",(i+1)," : ",mean)

row = np.mean(a,axis=1)
print(row)

column = np.mean(a,axis=0)
print(column)


# replace even by -1
even = np.where(a%2 == 0,-1,a)
print(even)

sort_row = np.sort(a,axis=1)
sort_col = np.sort(a,axis=0)
print(sort_row)
print(sort_col)


# row and column wise sum
sum_row = np.sum(a,axis=1)
sum_col = np.sum(a,axis=0)
print(sum_row)
print(sum_col)


# number of "-ve" in each column
count_col = np.sum(np.where(even<0,1,0),axis=0)
print(count_col)

```

```python
stock_level = np.random.randint(20,80,size=(5,5))
demand = np.random.randint(20,80,size=(5,5))
print(stock_level)
print(" ")
print(demand)
print(" ")


# remaining stocks
remaining_stock = stock_level - demand
print(remaining_stock)
print(" ")


# replace negative eith zero
replace_negative = np.where(remaining_stock<0,0,remaining_stock)
print(replace_negative)
print(" ")



# total shortage
total_shortage = np.sum(np.where(remaining_stock<0,remaining_stock,0))
print("total_shortage : ",total_shortage*(-1))
print(" ")



# product with zero stock
product_with_zero = np.sum(np.where(remaining_stock<=0,1,0),axis=0)
print("product_with_zero_stock : ",product_with_zero)
print(" ")



# sort each row on remaining stock
sort_row = np.sort(replace_negative,axis=1)
sort_col = np.sort(replace_negative,axis=0)
print(sort_row)
print(sort_col)
```
