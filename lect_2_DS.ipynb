```
# list
lst = [1,2,3,"hello",'c',[2,4]]
print(lst)


# append()
lst.append([5,6,7])
lst.append(2)
print(lst)


# extend()
# it append single element not a list
lst.extend([5,6,7])
print(lst)


# insert()
# lst.insert(index,value)
lst.insert(2,"value")
print(lst)


# pop()
lst.pop()
print(lst)


# remove()
# it removes the first occurence of the value
lst.remove(2)
print(lst)


# index()
# gives the index of the value
print(lst.index("value"))


# count()
print(lst.count(2))


# sort()
lt = [1,2,3,6,4,2,4,5]
lt.sort()
print(lt)


# sum()
sum(lt)
```

```
# sum min max
lst = [2,3,5,4,3,0,-1,9]
print(sum(lst))
print(min(lst))
print(max(lst))


# sorted return a brand new list without modifying the original one
print(sorted(lst))
print(lst)
```

```
# sum of two list
n = int(input())
a = []
b = []
for i in range(n):
  new = int(input())
  a.append(new)
for i in range(n):
  new = int(input())
  b.append(new)

result = []
for i in range(n):
  sm = a[i] + b[i]
  result.append(sm)

print(result)



# second method
n = int(input())
a = input("list 1:").split(" ")
b = input("list 2:").split(" ")
result = []
for i in range(n):
  result.append(int(a[i]) + int(b[i]))

print(result)
```
