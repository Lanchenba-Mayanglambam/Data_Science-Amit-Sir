```python
import numpy as np
```

```python
# creating an array using numpy

arr = np.array([1,3,9,0,2])
print(arr)



# if there is a different data type then numpy array will convert all the element into a single datatype
arr1 = np.array([1,2,3,4,5,'9'])
print(arr1)


# QUESTION
np_arr = np.array([3,3,3,3,3.3,'hello',False])
print(bool(np_arr[-1]))
print(np_arr)

a = False
print(bool(a))
```

```python
# zeros() function and it return aa numpy float by default to change the return datatype we use dtype
# zeros(3) creates a list(numpy array) of 0s
a = np.zeros(3)
print(a)

# zeros() functions as 2D Array
b = np.zeros((2,3))
print(b)

# multiple array
c = np.zeros((2,3,4),dtype=int)
print(c)
print(type(c[0][0][0]))
```

```python
# empty() function creates a numpy array of garbage values
a = np.empty(4,dtype=int)
print(a)
```

```python
# full() function creates a full numpy array of given value
a = np.full(4,5,dtype=int)
print(a)
```

```python
# arange() function gives a range of element starting from 0 upto n-1   syntax -> (start,end,jump)
a = np.arange(0,5,1)
print(a)
```

```python
# linspace() function  gives a numpy array  syntax -> (start,end,number of element want)
a = np.linspace(0,5,5,dtype=int)
print(a)
```

```python
# shape and size
a = np.zeros((2,3),dtype=int)
print(a.size)
print(a.shape)
```

```python
# random.randomint() function and random.choice() function

# random.randit() gives random values syntax -> (start,stop,number of values want)
a = np.random.randint(5,20,5)
print(a)

# random.choice() gives random values from a given array  syntax -> (given array,number of values)
b = np.random.choice([1,3,5,7,9],2)
print(b)
```

```python
# Operations
a = np.array([1,5,3])
b = np.array([2,4,9])
print(a*2)
print(a+b)
print(a/b)
print(a%b)



# Operations doesn't work on list
c = [1,2,3,4,5]
print(c*2)       #it will print c twice
d = [2,4,1,3,6]
# print(c*d)
```

```python
# reshape() function change the shape of the array
a = np.random.randint(2,15,6)
print(a)
print(" ")

b = a.reshape(2,3)
print(b)
print(" ")

c = b.transpose()
# print(c)

d = np.concatenate(b)
print(d)
```

```python
# concatenate() function
# a = np.random.randint(1,10,5)
# b = np.random.randint(1,10,5)
a = np.array([[1,2,3],[4,5,6]])
b = np.array([[7,8,9],[10,11,12]])

# print(a)
# print(b)
print(np.concatenate((a,b),axis=1))
```

```python
# where() function  syntax -> (condition,(true then this,false then this))
a = np.random.randint(1,10,5)
b = np.random.randint(1,10,5)

print(a)
print(b)

c = np.where(b>2,b,a)
print(c)
```

```python
# sort() function
a = np.random.randint(1,10,5)
print(a)
a.sort()
print(a)
```

```python
# sum() function
a = np.random.randint(1,10,5)
b = np.random.randint(1,10,5)

print(a)
print(b)

c = np.sum(a)
d = b.sum()
print(c)
print(d)
```

```python
# mean() function
a = np.random.randint(1,10,5)
print(a)

b = np.mean(a)
print(b)
```

```python
# std() function
a = np.random.randint(1,10,5)
print(a)

b = np.std(a)
print(b)
```

```python
# unique() function return an array of unique elements
a = np.random.randint(1,10,5)
print(a)

b = np.unique(a)
print(b)
```

```python
# matmul() function
# a = np.random.randint(1,10,5)
a = np.array([[1,2,3],[4,5,6]])
b = np.array([[1,2],[3,4],[5,6]])
# b = np.random.randint(1,10,5)

print(a)
print(b)

c = np.matmul(a,b)
print(c)
```

```python
# practice
n = 9
score = np.zeros((n),dtype=int)
for i in range(n):
  a = input(int())
  score[i] = a

for i in range(n):
  print("On day ",(i+1)," ","Player score : ",score[i]," runs")

x = np.max(score)
print("Maximum Score of the Player : ",x," run")

y = np.min(score)
print("Minimum Score of the Player : ",y," run")
```

```python
# taking input using split() function split convert all the elements into string
scores = input().split()
print(scores)


# to convert string into int
scores = [int(x) for x in scores]
print(scores)
```
